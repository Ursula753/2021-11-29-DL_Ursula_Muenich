```python
import pandas as pd
df = pd.read_csv("../data/Library_Usage.csv")
df.columns
```

```python
# wie viele Kinder sind Kunden?
print(len(df[df['Age Range'] == '0 to 9 years']))
print(len(df[df['Age Range'] == '10 to 19 years']))
```

```python
# wie viele Senioren sind Kunden?
print(len(df[df['Age Range'] == '65 to 74 years']))
```

```python
# Wie viele Nutzer möchten per eMail informiert werden?
print(len(df[df['Notice Preference Code'] == 'z'])).sum()
```

```python
# wie alt sind diese Kunden durchschnittlich?
df[['Age Range', 'Provided Email Address']].groupby(by='Age Range').mean()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
sns.set()

# use sample to generate a random subsample
df = pd.read_csv("../data/Library_Usage.csv").sample(n=1000)
```

```python
sns.catplot(y='Age Range', hue='Provided Email Address', 
            data=df, kind='count')
```

```python
sns.catplot(y='Age Range', hue='Notice Preference Definition', 
            data=df, kind='count') 
```

```python
sns.catplot(x='Notice Preference Definition', 
            data=df, kind='count', 
            col='Age Range', col_wrap=4)
```

```python
sns.barplot(x='Notice Preference Definition',
            y="Age Range", 
            data=df, 
            order=df.sort_values('Age Range').Notice Preference Definition)
```
