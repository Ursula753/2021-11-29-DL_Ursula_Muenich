```python
url ="https://www.bibsonomy.org/json/search/database?items=20"
```

```python
f = urllib.request.urlopen(url)
```

```python
result = f.read().decode('utf-8')
```

```python
print(type(result))
```

```python
data = json.loads(result)
```

```python
print(json.dumps(data, indent=2))
```

```python
edited_items = []

# Auswahl der Informationen, die ausgewhlt werden sollen:

desired_entries = ['type', 'year', 'publisher', 'author', 'label', 'url']
```

```python
data.keys()
```

```python
for item in data["items"]:

    # erstelle ein neues leeres dictionary fr die Daten des Elements

    item_export = {}
    #   and len(item["user"])<4
    # fr jeden gewnschten Eintrag
    # hier knnen Filterkriterien hinzugefgt werden (hier z.B. erschienen nach 2009 und mit weniger als 4 Usern):
    #if int(item["year"])>2009:
    for entry in desired_entries:
            # ... berprfe, ob es einen Eintrag gibt (um Fehler zu vermeiden)
        if entry in item.keys():
                # fge den Eintrag den Daten zum Element hinzu
            item_export[entry] = item[entry]
        # am Ende: fge den vollstndigen (gefilterten) Eintrag der zu exportierenden Liste hinzu
    print(edited_items)
    edited_items.append(item_export)
```
